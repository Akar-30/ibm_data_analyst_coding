```python
import pandas as pd
import sqlite3
conn = sqlite3.connect('FinalDB.db')
census_df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoCensusData.csv')
public_schools_df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoPublicSchools.csv')
crime_df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DB0201EN-SkillsNetwork/labs/FinalModule_Coursera_V5/data/ChicagoCrimeData.csv')
census_df.to_sql('CENSUS_DATA', conn, if_exists='replace', index=False, method='multi')
```

```python
public_schools_df.to_sql('CHICAGO_PUBLIC_SCHOOLS', conn, if_exists='replace', index=False )
```

### Problem 8

##### Which community area is most crime prone? Display the coumminty area number only.



```python
%sql SELECT "COMMUNITY_AREA_NUMBER", count(*) as "crime_count" FROM CHICAGO_CRIME_DATA GROUP BY "COMMUNITY_AREA_NUMBER" HAVING count(*) = (SELECT MAX(crime_count) FROM (SELECT count(*) as crime_count FROM CHICAGO_CRIME_DATA GROUP BY "COMMUNITY_AREA_NUMBER"));
```
