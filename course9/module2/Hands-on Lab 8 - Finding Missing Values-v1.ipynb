##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
# Write your code here
# Create a boolean DataFrame where True indicates missing values
df_missing_values = df.isnull()

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(df_missing_values, 
            cbar=True, 
            cmap='viridis',
            yticklabels=False,  # Hide row labels for cleaner view
            xticklabels=True)   # Show column names

plt.title('Missing Values Heatmap', fontsize=16, fontweight='bold')
plt.xlabel('Columns', fontsize=12)
plt.ylabel('Rows', fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotate column names for better readability
plt.tight_layout()
plt.show()

# Additional summary of missing values
print("Summary of Missing Values:")
print("-" * 30)
missing_summary = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) * 100

for col in df.columns:
    if missing_summary[col] > 0:
        print(f"{col}: {missing_summary[col]} ({missing_percentage[col]:.2f}%)")
    else:
        print(f"{col}: No missing values")
```
