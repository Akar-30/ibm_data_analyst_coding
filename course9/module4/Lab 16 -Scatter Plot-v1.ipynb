#### 2. Scatter Plot for Popular Programming Languages by Job Satisfaction


Visualize the popularity of programming languages (`LanguageHaveWorkedWith`) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.


```python
## Write your code here
unique_values_in_worked_with_lang = df['LanguageHaveWorkedWith'].unique()
unique_items = set(x for i in unique_values_in_worked_with_lang for x in str(i).split(';'))
unique_items
```

```python
lang_worked_with = dict()
for item in unique_items:
    lang_worked_with[item] = df['LanguageHaveWorkedWith'].str.contains(item, regex=False).sum()
lang_worked_with
```

```python
df['LanguageHaveWorkedWith'].str.contains('Bash/Shell (all shells)',regex=False).sum()
```

```python
# Create a DataFrame from the language popularity dictionary
lang_popularity_df = pd.DataFrame(list(lang_worked_with.items()), columns=['Language', 'Popularity'])

# Remove languages with 0 popularity and 'nan' entries
lang_popularity_df = lang_popularity_df[(lang_popularity_df['Popularity'] > 0) & (lang_popularity_df['Language'] != 'nan')]

# Get average job satisfaction for each language
lang_satisfaction = {}
for lang in lang_popularity_df['Language']:
    # Create boolean mask for users who worked with this language
    mask = df['LanguageHaveWorkedWith'].str.contains(lang, na=False)
    # Calculate average job satisfaction for users of this language
    avg_satisfaction = df[mask]['JobSatPoints_6'].mean()
    lang_satisfaction[lang] = avg_satisfaction

# Add satisfaction data to the dataframe
lang_popularity_df['AvgJobSatisfaction'] = lang_popularity_df['Language'].map(lang_satisfaction)

# Remove any rows with NaN satisfaction values
lang_popularity_df = lang_popularity_df.dropna()

# Create scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(data=lang_popularity_df, x='AvgJobSatisfaction', y='Popularity', s=100, alpha=0.7)

# Add labels for some popular languages
for i, row in lang_popularity_df.iterrows():
    if row['Popularity'] > 10000 or row['Language'] in ['Python', 'JavaScript', 'Java', 'C++', 'C#']:
        plt.annotate(row['Language'], (row['AvgJobSatisfaction'], row['Popularity']), 
                    xytext=(5, 5), textcoords='offset points', fontsize=9)

plt.xlabel('Average Job Satisfaction')
plt.ylabel('Language Popularity (Number of Users)')
plt.title('Programming Language Popularity vs Job Satisfaction')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

### Task 4: Scatter Plot Comparisons Across Groups


#### 1. Scatter Plot for Compensation vs. Job Satisfaction by Employment Type


Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), categorized by employment type (`Employment`). Use color coding or markers to differentiate between employment types.


```python
unique_values_in_employment = df['Employment'].unique()
unique_items_employment = set(x for i in unique_values_in_employment for x in str(i).split(';'))
unique_items_employment
```

```python
plt.figure(figsize=(16, 12))
markers = ['o', 's', 'D', '^', 'p', '*', 'h', 'H', 'X', 'd']
for i, marker in zip(unique_items_employment, markers):
    sns.scatterplot(data=df[df['Employment'].str.contains(i, na=False)], y='ConvertedCompYearly', x='JobSatPoints_6', label=i, marker=marker)
```

#### 2. Scatter Plot for Work Experience vs. Age Group by Country


Compare work experience (`YearsCodePro`) across different age groups (`Age`) and countries (`Country`). Use colors to represent different countries and markers for age groups.


```python
## Write your code here

# Get top countries by count for better visualization
top_countries = df['Country'].value_counts().head(10).index.tolist()

# Create age groups
age_groups = ['Under 18 years old', '18-24 years old', '25-34 years old', '35-44 years old', '45-54 years old', '55-64 years old', '65 years or older']
markers = ['o', 's', 'D', '^', 'v','*', 'h',]

plt.figure(figsize=(14, 10))

# Create a color palette for countries
colors = plt.cm.viridis(np.linspace(0, 1, len(top_countries)))

for i, country in enumerate(top_countries):
    for j, age_group in enumerate(age_groups):
        # Filter data for specific country and age group
        mask = (df['Country'] == country) & (df['Age'] == age_group)
        subset = df[mask]
        
        if len(subset) > 0:
            plt.scatter(subset['YearsCodePro'], subset['Age_numeric'], 
                       color=colors[i], marker=markers[j], 
                    #    label=f'{country} - {age_group}' if i == 0 else '', 
                       alpha=0.35, s=50)

# Create custom legends
country_legend = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=colors[i], 
                            markersize=8, label=country) for i, country in enumerate(top_countries)]
age_legend = [plt.Line2D([0], [0], marker=markers[j], color='gray', 
                        markersize=8, label=age_group) for j, age_group in enumerate(age_groups)]

# Add legends
legend1 = plt.legend(handles=country_legend, title='Countries', loc='upper left', bbox_to_anchor=(1.05, 1))
legend2 = plt.legend(handles=age_legend, title='Age Groups', loc='upper left', bbox_to_anchor=(1.05, 0.6))
plt.gca().add_artist(legend1)

plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Age (Numeric)')
plt.title('Work Experience vs Age by Country and Age Group')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
