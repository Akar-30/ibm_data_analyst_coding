```python
df['WorkedWith_Bash/Shell (all shells)'] = df["LanguageHaveWorkedWith"].str.contains('Bash/Shell (all shells)', regex=False).astype("Int64")
df[['WorkedWith_Bash/Shell (all shells)']]
```

```python
for i in df.columns:
    print(i) if i.startswith('WorkedWith_') else None
```

```python
df_grouped = df.groupby('Age_numeric')[[i for i in df.columns if i.startswith('WorkedWith_')]].sum().reset_index()
df_grouped
```

#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
plt.figure(figsize=(10, 8))
for i in df_grouped.columns:
    if i.startswith('WorkedWith_'):
        sns.scatterplot(data=df_grouped, y='Age_numeric', x=i, size=i, sizes=(20, 200), alpha=0.5, legend=False)
plt.title('Bubble plot of Age vs. Technology Worked With')
plt.legend(title='Technology Worked With', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
sns.scatterplot(data=df_grouped, y='Age_numeric', x='WorkedWith_Bash/Shell (all shells)', size='WorkedWith_Bash/Shell (all shells)', sizes=(20, 200), alpha=0.5, legend=True)
plt.legend(title='WorkedWith_Bash/Shell (all shells)', bbox_to_anchor=(1.05, 1), loc='upper left')
```
